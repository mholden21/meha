```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("exoplanet_data.csv", comment = "#")
```

```python
#Planet Name
#pl_name
#planet name most commonly used in literature
#this parameter gives the name of the planet
```

```python
#host name
#hostname
#Stellar name most commonly used in the literature
#the name of the stellar object (star) the exoplanet orbits/is hosted by
```

```python
#discovery method
#discoverymethod
#Method by which the planet was first identified
#the reason the scientists discovered the exoplanet
```

```python
#spectral type
#st_spectype
#Classification of the star based on their spectral characteristics following the Morgan-Keenan system
#classifying stars based on temperature with hottest O, B, A, F, G, K, M, L to coldest
```

```python
#equilibrium temperature
#pl_eqt
#The equilibrium temperature of the planet as modeled by a black body heated only by its host star, or for directly imaged planets, the effective temperature of the planet required to match the measured luminosity if the planet were a black body
#state where absorbed energy equals emitted energy (acts as a blackbody)
```

```python
#planet mass or mass*sin(i) [jupiter mass]
#pl_bmassj
#Minimum mass of a planet as measured by radial velocity, measured in units of masses of Jupiter
#mass measured using the "wobble" of the star in its orbit (changes in radial velocity) represented in units of the mass of jupiter
```

```python
df2 = df[["pl_name", "hostname", "discoverymethod", "st_spectype", "pl_eqt", "pl_bmassj"]].copy()
```

```python
df3 = df2.dropna()
```

```python
df3.shape
#67 rows, 6 columns
df3.head()
df3.tail()
print(df3.columns)
```

```python
df3["discoverymethod"].value_counts()
```

```python
#different discovery methods are transit, radial velocity, and imaging
#the most common method is transit with 52, while the least is imaging with only 2. Radial velocity is in the middle with 13
#radial velocity is when exoplanets are spotted based on changes in a sun's motion
#imaging uses telescopes to capture faint light of planets as they orbit their host star
#transit involved finding exoplanets through noting a dip in the stars brighness as the exoplanet orbits the star
```

```python
plt.figure(figsize= (8,6))
sns.scatterplot(data=df3, x="pl_bmassj", y="pl_eqt", hue="discoverymethod")
plt.title("Mass vs Temperature by discovery method")
plt.xlabel("mass (in units of mass of jupiter)")
plt.ylabel("temperature in kelvin")
plt.legend(title="discovery method")
plt.show()
```

```python
plt.figure(figsize= (8,6))
sns.scatterplot(data=df3, x=np.log(df3["pl_bmassj"]), y=np.log(df3["pl_eqt"]), hue="discoverymethod")
plt.title("Mass vs Temperature log transformed by discovery method")
plt.xlabel("mass (in units of mass of jupiter)")
plt.ylabel("temperature in kelvin")
plt.legend(title="discovery method")
plt.show()
```

```python
#exoplanets found by imaging are more likely to be massive
heavy = df3[df3["pl_bmassj"]>5]
cold_heavy = heavy.loc[heavy["pl_eqt"].idxmin()]
print([heavy["pl_eqt"].idxmin()])
print(cold_heavy)
```

```python
#the planet name is eps Ind A b
#the planet index is 566
#the equilibrium temperature is 275 kelvin
#its mass is 6.31 jupiters
#its orbiting star eps Ind A, with spectral type K5 V, unlike the sun which is G2V
```

```python
plt.figure(figsize= (8,6))
sns.scatterplot(data=df3, x=np.log(df3["pl_bmassj"]), y=np.log(df3["pl_eqt"]), hue="discoverymethod")
plt.scatter(np.log(cold_heavy["pl_bmassj"]), np.log(cold_heavy["pl_eqt"]), color = "yellow", marker = "*", s = 200, label = "eps Ind A b")
plt.title("Mass vs Temperature log transformed by discovery method")
plt.xlabel("log mass (in units of mass of jupiter)")
plt.ylabel("log temperature in kelvin")
plt.legend(title="discovery method")
plt.show()
```
