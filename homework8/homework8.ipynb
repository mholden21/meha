```python
print(df["latitude"].median())
print(df["longitude"].median())
```

```python
median = df["latitude"].median()
df["latitude_to_median"] = (df["latitude"] - median).abs()
median2 = df["longitude"].median()
df["longitude_to_median"] = (df["longitude"] - median2).abs()
print(df)
```

```python
print(df.shape)
```

```python
close_row = df.loc[df["latitude_to_median"].idxmin()]
print(f"closest to median latitude: {close_row["state"]} by {close_row["latitude_to_median"]:.2f} degrees")
```

```python
far_row = df.loc[df["latitude_to_median"].idxmax()]
print(f"farthest to median latitude: {far_row["state"]} by {far_row["latitude_to_median"]:.2f} degrees")
```

```python
close_row2 = df.loc[df["longitude_to_median"].idxmin()]
print(f"closest to median longitude: {close_row["state"]} by {close_row2["longitude_to_median"]:.2f} degrees")
```

```python
far_row2 = df.loc[df["longitude_to_median"].idxmax()]
print(f"farthest to median longitude: {far_row2["state"]} by {far_row2["longitude_to_median"]:.2f} degrees")
```

```python
import seaborn as sns
```

```python
df3 = sns.load_dataset("flights")
```

```python
df3.dropna(axis=1)
```

```python
df3.shape
#144 rows, 3 columns
```

```python
print(df3.head())
print(df3.tail())
```

```python
#the columns represent the year the flight was taken, the month, and the passengers
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 12))
axes[0,0].plot(df3["year"], df3["month"], color = "blue")
axes[0,0].set_title("Year vs Month of flight")
axes[0,0].set_xlabel("years")
axes[0,0].set_ylabel("months")

axes[0,1].plot(df3["month"], df3["passengers"], label = "travel based on month", color = "red")
axes[0,1].set_title("Months vs Passengers flights")
axes[0,1].set_xlabel("months")
axes[0,1].set_ylabel("passengers")
axes[0,1].legend()

sns.histplot(data= df3, x = "passengers", ax= axes[1,0], color = "green")
axes[1,0].set_title("histogram of passengers")
axes[1,0].set_xlabel("passengers")

sns.boxplot(data = df3, x = "month", ax= axes[1,1], color = "yellow")
axes[1,1].set_title("boxplot of months")
axes[1,1].set_xlabel("months")
```

```python
#I was a bit restricted by the limited number of columns in the dataset,but the graphs reveal some interesting information
#the months vs passengers shows that travel spikes in the summertime, from june-august
#the passengers histogram reveals that most flights take 200 passengers, while 600 passenger flights are very rare
```
