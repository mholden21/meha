```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv("noisy_curvefit_data.csv")
```

```python
print(df.shape)
#100 rows, 6 columns
print(df.head())
print(df.tail())
x = df["x"]
y_noise_1 = df["y_noise_1"]
#columns names are x, y noise 1, y noise 2, y noise 3, y noise 4, and y noise 5
```

```python
plt.scatter(df["x"], df["y_noise_1"], label = "data")
plt.title("X vs Y noise 1")
plt.xlabel("X")
plt.ylabel("y noise 1")
plt.legend()
plt.show()
```

```python
def model (x,A,k):
    return A * np.exp(-k * x)
p0 = [1,1]
popt, pocv = curve_fit(model, x, y_noise_1, p0=p0)
A_fit, k_fit = popt
f"Fit: y = {A_fit:.2f} * e^(-{k_fit:.2f} * x)"
residuals = y_noise_1 - model(x, A_fit, k_fit)
chi2 = np.sum(residuals**2)
dof = len(x) - len(popt)
chi2_reduced = chi2 / dof
plt.scatter(df["x"], df["y_noise_1"], label = "data")
plt.plot(x, model(x, A_fit, k_fit,), color = "red", label = "Fitted model")
plt.plot(x, model(x, 5, 0.8), color = "black", linestyle = "dotted", label = "true model")
plt.title(f"Exponential Fit Y Noise 1, Reduced  = {chi2_reduced:.2f})")
plt.legend()
plt.xlabel("x values")
plt.ylabel("y noise 1 values")
```

```python
def fit_model(x,y):
    p0 = [1,1]
    popt, pocv = curve_fit(model, x, y, p0=p0)
    A_fit, k_fit = popt
    residuals = y - model(x, A_fit, k_fit)
    chi2 = np.sum(residuals**2)
    dof = len(x) - len(popt)
    chi2_reduced = chi2 / dof
    return popt, chi2_reduced
    
fig, axes = plt.subplots(2, 2, figsize = (8,10))
axes = axes.flatten()

columns = ["y_noise_2", "y_noise_3", "y_noise_4", "y_noise_5"]
noise_levels = [2, 3, 4, 5]
for i, (columns2, noise) in enumerate(zip(columns, noise_levels)):
    y = df[columns2]
    popt, chi2 = fit_model(x, y)
    A_fit, k_fit = popt
    ax = axes[i]
    ax.scatter(x, y, color='blue', alpha=0.6, label='Data')
    ax.plot(x, model(x, A_fit, k_fit), color='red', label='Fitted model')
    ax.plot(x, model(x, 5, 0.8), color = "black", linestyle = "dotted", label = "true model")
    ax.set_title(f"Fit Noise {noise}, Reduced  = {chi2:.2f}")
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()

    
```

```python
#as you increase the noise level, the red and black dotted line overlap less
#as you increase the noise level, the chi squared number gets farther and farther away from 1, 
#which means the line of best fit model represents the data less accurately the higher the nosie level is 
```

```python
columns = ["y_noise_1", "y_noise_2", "y_noise_3", "y_noise_4", "y_noise_5"]
noise_levels = [1, 2, 3, 4, 5]
results = []

for col, noise in zip(columns, noise_levels):
    y = df[col]
    popt, chi2 = fit_model(x, y)
    A_fit, k_fit = popt
    results.append({
        "Noise level": noise,
        "A fit": A_fit,
        "k fit": k_fit,
        "Reduced chi-squared": chi2
    })

new_df = pd.DataFrame(results)
print(new_df)
```

```python
new_df.to_csv("noisy.parameters.csv", index = False)
```
