```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv("solar_system.csv")
```

```python
print(df.shape)
#shape is 20 rows, 11 columns
df.head()
```

```python
df.tail()
```

```python
print(df.columns)
print(df["Attribute"])
```

```python
orbital_period = df[df["Attribute"] == "Orbital Period (days)"].iloc[0, 1:].astype(float)
print(orbital_period)
```

```python
perihelion = df[df["Attribute"] == "Perihelion (10^6 km)"].iloc[0, 1:].astype(float)
aphelion = df[df["Attribute"] == "Aphelion (10^6 km)"].iloc[0, 1:].astype(float)
```

```python
#orbital period is the period in days of how long the planet takes to complete a full rotation aroud the sun
#the perihelion is the point in the orbit where the planet is closest to the sun
#aphelion is the point in the orbit where the planet is farthest from the sun
```

```python
semi_major_axis = (perihelion + aphelion) / 2
new_row = pd.DataFrame({
    "Attribute": ["Semi-Major Axis (AU)"],
    **{planet: [value] for planet, value in semi_major_axis.items()}
})
df = pd.concat([df, new_row], ignore_index = True)
```

```python
print(df)
```

```python
df.shape
#21 rows, 11 columns
```

```python
plt.figure(figsize=(7,5))
plt.scatter(semi_major_axis, orbital_period, label = "data")
plt.xlabel("Semi-major axis (AU)")
plt.ylabel("Orbital period (days)")
plt.title("semi major axis vs orbital period")
plt.legend()
plt.show()
```

```python
x = semi_major_axis
y = orbital_period
def keplers_third_law(a, m, k):
    return m * a ** k
p0 = [1,1.5]
popt, pcov = curve_fit(keplers_third_law, x, y, p0=p0)
A_fit, k_fit = popt

residuals = y - keplers_third_law(x, A_fit, k_fit)
chi2 = np.sum(residuals**2)
dof = len(x) - len(popt)
chi2_reduced = chi2 / dof
print(f"Fit: y = {A_fit:.2f} * a^(-{k_fit:.2f})")

plt.figure(figsize=(7,5))
plt.scatter(x, y, label = "data")
plt.plot(x, keplers_third_law(x, A_fit, k_fit,), color = "red", label = "Fitted model")
plt.title(f"Kepler's Model Fit, Reduced  = {chi2_reduced:.2f}")
plt.legend()
plt.xlabel("semi major axis (AU)")
plt.ylabel("orbital period (days)")
```

```python
df.to_csv("kepler.csv", index = False)
```
