```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("homework7/star_data.txt", skiprows=1)
```

```python
print(data.shape)
#2642 rows
```

```python
magnitude = data[:,0]
temperature = data[:,1]

mag = magnitude
temp = temperature

print(mag.max())
print(mag.min())
print(mag.argmax())
print(mag.argmin())

print(temp.max())
print(temp.min())
print(temp.argmax())
print(temp.argmin())
```

```python
# 6. star 1988 is the brightest, while star 264 is the dimmest
```

```python
log_mag = np.log(mag)
log_temp = np.log(temp)

plt.figure(figsize=(8,6))
plt.scatter(log_temp, log_mag, label = "stars")
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature")
plt.ylabel("Magnitude")
plt.scatter(
    log_temp[1988], 
    log_mag[1988],
    color = 'black',
    marker = '*',
    s = 300,
    label = "brightest star"
)
plt.scatter(
    log_temp[264],
    log_mag[264],
    color = 'purple',
    marker = '*',
    s = 300,
    label = "dimmest star"
)

plt.scatter(
    log_temp[2613],
    log_mag[2613],
    color = "green",
    marker = "*",
    s = 300,
    label = "hottest star"
)

plt.scatter (
    log_temp[1923],
    log_mag[1923],
    color = "orange",
    marker = "*",
    s = 300,
    label = "coldest star"
)

plt.legend()
plt.show
```
